## 5. Consistency Checks

#### Mixed-type columns

```python
# Search dataframe for mixed-type columns
for col in df_cust_del.columns.tolist():
  weird = (df_cust_del[[col]].map(type) != df_cust_del[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust_del[weird]) > 0:
    print (col)
```

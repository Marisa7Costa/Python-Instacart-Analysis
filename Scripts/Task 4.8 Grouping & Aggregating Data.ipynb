## 2. & 3. Aggregated mean of the “order_number” column grouped by “department_id”

## 4. Creating a loyalty flag for existing customers

```python
#Identify Loyal cutomers - group orders by user to generate max orders per user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#Check output
ords_prods_merge.head()
```

```python
#Create flag for max_order based on criteri given
#Cell 1 of criteria
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
#Cell 2 of criteria
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
#Last cell of criteria
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Check output
ords_prods_merge[['user_id','loyalty_flag','order_number']].head()
```

```python
#Run frequency count on max_order flag
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

## 5. Check the basic statistics of the product prices for each loyalty category

```python
#Mean, minimum, maximum and median stats per loyalty_flag based on prices
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean','min','max','median']})
```

It seems that loyal customers spend on average the least per product and new customers spend the most.
The maximum value shows a product price that needs to be looked at.

## 6. Create a spending flag for each user based on the average price across all their orders

```python
#The mean of the prices of products purchased by a user
ords_prods_merge['avg_order_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge[['user_id','order_number','prices','avg_order_price']].head()
```

```python
#Create a flag for avg_order_price based on critera given
#Cell 1 of criteria
ords_prods_merge.loc[ords_prods_merge['avg_order_price'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
#Last cell of criteria
ords_prods_merge.loc[(ords_prods_merge['avg_order_price'] >= 10), 'spending_flag'] = 'High Spender'
```

```python
#Check output
ords_prods_merge[['user_id','spending_flag','avg_order_price']].head(100)
```

```python
#Run frequency count on spending_flag
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

## 7. Order frequency flag

```python
#The median in the “days_since_prior_order” column
ords_prods_merge['order_behavior'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#check output
ords_prods_merge[['user_id','days_since_prior_order','order_behavior']].head()
```

```python
#Create a flag for order_behavior based on criteria given
#Cell 1 of criteria - greater than 20 or equal to nan (consumers who have only made one order)
ords_prods_merge.loc[(ords_prods_merge['order_behavior'] > 20) , 'order_frequecy'] = 'Non-frequent customer'
```

```python
#Cell 2 of criteria - higher than 10 and lower than or equal to 20
ords_prods_merge.loc[(ords_prods_merge['order_behavior'] <=20) & (ords_prods_merge['order_behavior'] > 10), 'order_frequecy'] = 'Regular customer'
```

```python
#Last cell of criteria - lower than or equal to 10
ords_prods_merge.loc[(ords_prods_merge['order_behavior'] <= 10), 'order_frequecy'] = 'Frequent customer'
```

```python
#Check output
ords_prods_merge[['user_id','order_behavior','order_frequecy']].head(100)
```

```python
#Run frequency count on order_frequency
ords_prods_merge['order_frequecy'].value_counts(dropna = False)
```
